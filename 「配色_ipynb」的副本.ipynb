```
install.packages("ggplot2")
```

```
library(SportsAnalytics)
library(ggplot2)
```

```
NBA1516 <-fetch_NBAPlayerStatistics("15-16")
qplot(PersonalFouls, TotalPoints, data=NBA1516, color=Position) #X=個人犯規 Y=總得分
```

```
NBA1516 <-fetch_NBAPlayerStatistics("15-16")
qplot(PersonalFouls, TotalPoints, data=NBA1516, geom=c("point", "smooth")) #加上漸進線
```

```
NBA1516 <-fetch_NBAPlayerStatistics("15-16")
qplot(FieldGoalsMade, data=NBA1516, fill=Position) #投籃進球數直方圖
```

```
qplot(PersonalFouls , TotalPoints,
    data = NBA1516, 
    facets = Position ~.) #X=個人犯規 Y=總得分 橫向散佈圖
```

```
ggplot( data = NBA1516, aes(x=Position , y =Steals )) + geom_boxplot() #X=位置 Y=抄球 盒狀圖
```
